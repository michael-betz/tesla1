```python
acc = 0  # simulated 32 bit unsigned
```

```python
BITS_PER_SEC =  1000000  # 1 Mbits/s
WORDS_PER_SEC = (BITS_PER_SEC / 32)
```

```python
f = 83724 # Hz

ftw = 2**32 * f // BITS_PER_SEC
print('  ftw = {:08x}, f_norm = {:.6f} f_s'.format(ftw, f / BITS_PER_SEC))

duty = 0.3
duty_ = int((2**32 - 1) * duty)
print('duty_ = {:08x},   t_on = {:.3f} clocks'.format(duty_, duty_ // ftw))

phase = int((2**32 - 1) * 0.7)

for i in range(2**12):
    acc = (acc + ftw) & 0xFFFFFFFF
    tmp = (acc + phase) & 0xFFFFFFFF
    print('*' if tmp < duty_ else '.', end='')
    if i % 72 == 0:
        print()
```
